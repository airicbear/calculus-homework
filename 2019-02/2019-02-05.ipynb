```julia
using Plots

trapezoid_area(h, a, b) = (h / 2) * (a + b)

plot_integral(f, a, b, label = "") = plot!(f.(a:b), fill = (0, 0.5), label = label)

function riemann(f, a, b, n)
    Δx = (b - a) / n;
    xs = @. a + (0:n-1) * Δx
    sum(f.(xs) * Δx)
end

function plot_integral_approximation(f, a, b, n)
    Δx = (b - a) / n;
    for i in 1+a:n+a
        Δxi = Δx*i-Δx:Δx/100:Δx*i
        y = f((i - 1) * Δx)
        plot!(Δxi, [y for _ in Δxi], fill = (0, 0.5), label = "")
    end
    plot!(fill = (0, 0.5))
end
```

### *Solution 1*

```julia
c(x) = -0.012x + 6.50
xs = 0:500
plot(xs, c, label = "c(x) = -0.012x + 6.50"
    , title = "Redline Roasting's cost of roasting 200lbs of coffee"
    , xlabel = "Pounds of coffee roasted"
    , ylabel = "Cost of roasting coffee in dollars per pound")
plot_integral(c, 0, 200, "Total cost of roasting 200lbs of coffee")
```

```julia
trapezoid_area(200, c(0), c(200))
```

### *Solution 5*

```julia
R′(x) = 2x - 1150
plot(0:1000, R′, label = "R'(x) = 2x - 1150"
    , title = "Concert promoter's total revenue from first 300 tickets sold"
    , xlabel = "Number of tickets sold"
    , ylabel = "Rate of change of total revenue")
plot_integral(R′, 0, 300, "Total revenue from the sale of the first 300 tickets")
```

```julia
riemann(R′, 0, 300, 100000)
```
