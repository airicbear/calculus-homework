$\text{Differentiate.}$

$1)$
$$\begin{align} \\
y &= xe^{-2x} + e^{-x} + x^3 \\
y' &= -2xe^{-2x} + e^{-2x} - e^{-x} + 3x^2 \\
&= -xe^{-2x} - e^{-x} + 3x^2
\end{align}$$

```python
y = lambda x: x * exp(-2 * x) + exp(-x) + power(x, 3)
dy = lambda x: -x * exp(-2 * x) - exp(-x) + 3 * power(x, 2)
x = linspace(0, 10, 100)
plot(x, y(x))
_ = plot(x, dy(x))
```

$\text{7) Find the equation of the line tangent to the graph of } f(x) = e^{-x} \text{ at the point (0, 1)}$
$$\begin{align} \\
&& f'(x) &= -e^{-x} \\
&& f'(0) &= -e^{0} \\
&& &= -1 \\
&& y &= mx + b \\
&& y &= f'(0)(x - 0) + f(0) \\
&& y &= -x + 1
\end{align}$$

```python
f = lambda x: exp(-x)
tangent_line = lambda x: -x + 1
x = linspace(-2, 2, 100)
plot(x, f(x))
_ = plot(x, tangent_line(x))
```

$\text{8) Find the equation of the line tangent to the graph of } f(x) = e^{3x} \cdot \ln{(4x)} \text{ at } x = 1$
$$\begin{align} \\
f'(x) &= e^{3x} \cdot \frac{4}{4x} + \ln{4x} \cdot 3e^{3x} \\
&= e^{3x} \left(\frac{1}{x} + 3\ln{4x}\right) \\
f'(1) &= e^{3(1)} \left(\frac{1}{(1)} + 3\ln{4(1)}\right) \\
&= e^3 \left(1 + 3 \ln{4}\right) \\
y &= f'(1)(x - 1) + f(1) \\
&= e^3 \left(1 + 3 \ln{4}\right)(x - 1) + e^3 \cdot \ln{4}
\end{align}$$

```python
f = lambda x: exp(3 * x) * log(4 * x)
def df(x):
    return exp(3 * x) * ((1 / x) + 3 * log(4 * x))
tangent_line = lambda x: df(1) * (x - 1) + f(1)
x = linspace(0.5, 1.5, 100)
plot(x, f(x))
_ = plot(x, tangent_line(x))
```
